Average Filtering


```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np


im= cv.imread(r"D:\Saves\Python\ImgPro\images\sigiriya.jpg", cv.IMREAD_REDUCED_GRAYSCALE_2)
assert im is not None

w = 9
kernel= np.ones((w,w),np.float32)/(w*w)

im2 = cv.filter2D(im, -1, kernel)

fig, ax= plt.subplots(1,2, figsize=(10,20))
ax[0].imshow(im, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Original')
ax[1].imshow(im2, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Average Filtered')
plt.show()
```

Comparison

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np


im= cv.imread(r"D:\Saves\Python\ImgPro\images\sigiriya.jpg", cv.IMREAD_REDUCED_GRAYSCALE_2)
assert im is not None


sobel_h= np.array([[-1,0,1],[-2,0,2],[-1,0,1]], np.float32)
imh = cv.filter2D(im.astype('float'), -1, sobel_h)
imh = cv.normalize(imh.astype('float'), None,0, 255.0, cv.NORM_MINMAX)

sobel_v= np.array([[-1,-2,-1],[0,0,0],[1,2,1]], np.float32)
imv = cv.filter2D(im.astype('float'), -1, sobel_v)
imv = cv.normalize(imv.astype('float'), None,0, 255.0, cv.NORM_MINMAX)

fig, ax= plt.subplots(1,3, figsize=(10,20))
ax[0].imshow(im, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Original')
ax[1].imshow(imh, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Sobel Horizontal Filtered')
ax[2].imshow(imv, cmap='gray', vmin=0, vmax=255)
ax[2].set_title('Sobel Vertical Filtered')
plt.show()
```

Sharpening Tom Cruise

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np


im= cv.imread(r"D:\Saves\Python\ImgPro\images\tom.jpg", cv.IMREAD_REDUCED_GRAYSCALE_2)
assert im is not None

blurred = cv.blur(im,(11,11))
hp = cv.subtract(im, blurred)
sharpened = cv.addWeighted(im,1.0,hp,1.5,0)



fig, ax= plt.subplots(1,4, figsize=(20,40))
ax[0].imshow(im, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Original')
ax[1].imshow(blurred, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Blurred')
ax[2].imshow(hp, cmap='gray', vmin=0, vmax=255)
ax[2].set_title('High Frequencies')
ax[3].imshow(sharpened, cmap='gray', vmin=0, vmax=255)
ax[3].set_title('Sharpened')
plt.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

y, x= np.arange(-2, 2.1, 0.1), np.arange(-2, 2.1, 0.1)
Y, X= np.meshgrid(x,y)
sigma = 1.
g = np.exp(-(X**2+Y**2)/(2*sigma**2))

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, g, cmap=cm.jet, linewidth=0, antialiased=True)
cset = ax.contourf(X, Y, g, zdir='z', offset=np.min(g) -1.5, cmap=cm.jet)
ax.set_zlim(np.min(g) - 2, np.max(g))
plt.axis('off')
```
